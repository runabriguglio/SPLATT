```python
def analyse_acc_list(wdflist, f_vec, fs=freq_WebDAQ, n_bins:int=0):

    peak_spe = np.zeros([len(wdflist),4])
    peak_fs = np.zeros([len(wdflist),4])

    dt = 1/fs
    for k, wdf in enumerate(wdflist):
        
        data = sp.openfile(wdf)
        Nch = np.shape(data)[0]
        ff = f_vec[k]

        for j in range(Nch):
            #filt = sp.butterworth_bandpass_filter(data[j], f_low=ff-15, f_high=ff+15, f_sample=fs)
            #filt = filt[thr:] # remove initial transient
            #spe, f = sp.get_spectrum(filt, dt=dt)
            spe, f = sp.get_spectrum(data[j], dt=dt)
            
            spe = spe[f < 130]
            f = f[f < 130]
            #spe = spe[f > np.maximum(10,ff-60)]
            #f = f[f > np.maximum(10,ff-60)]
            peak_id = np.argmax(spe)
            peak_ids = np.arange(n_bins*2+1) + peak_id
            peak_spe[k,j] = np.sum(spe[peak_ids])
            peak_fs[k,j] = f[peak_id]


    return peak_spe, peak_fs

spe, f = analyse_acc_list(wdflist0, freq_vec0)

plt.figure()
plt.plot(f,spe)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [g]'); plt.yscale('log')
plt.grid(True)
```

```python
def analyse_acc_list(wdflist, f_vec, fs=freq_WebDAQ, n_bins:int=0):

    peak_spe = np.zeros([len(wdflist),4])
    peak_fs = np.zeros([len(wdflist),4])

    dt = 1/fs
    for k, wdf in enumerate(wdflist):
        
        data = sp.openfile(wdf)
        Nch, Lsig = np.shape(data)
        ff = f_vec[k]
        Nsampl = fs/ff # number of samples for a full cycle
        N = 9 # number of windows
        w_size = int(Lsig/N) #int(np.round(np.minimum(Lsig/N,Nsampl*24))) # window size
        w_shift = int((Lsig-w_size)/(N-1))

        for j in range(Nch):
            mean_spe = []
            for ii in range(N):
                ids = np.arange(w_size) + w_shift*ii
                acc = data[j,ids]
                spe, f = sp.get_spectrum(acc, dt=dt)
                if ii == 0:
                    mean_spe = spe**2/N
                else:
                    mean_spe += spe**2/N
            
            mean_spe = np.sqrt(mean_spe)
            spe = mean_spe[f < 130]
            f = f[f < 130]
            #spe = spe[f > np.maximum(10,ff-60)]
            #f = f[f > np.maximum(10,ff-60)]
            peak_id = np.argmax(spe)
            peak_ids = np.arange(n_bins*2+1) + peak_id
            peak_spe[k,j] = np.sum(spe[peak_ids])
            peak_fs[k,j] = f[peak_id]


    return peak_spe, peak_fs

spe, f = analyse_acc_list(wdflist0, freq_vec0)
#spe1, f = analyse_acc_list(wdflist1, freq_vec1, n_bins=2)

plt.figure()
plt.plot(f,spe,'-x')
#plt.plot(f,spe1,'--')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [g]')
plt.legend(('OBB','Stand','Piezo','Spring'))
plt.yscale('log')
plt.grid(True)

#print(f)
```
